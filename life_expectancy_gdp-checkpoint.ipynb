```python
#Data Collection
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


gdp_data = pd.read_csv('all_data.csv')
print(gdp_data.head())
```

```python
#Data Cleaning and Preparation 

#numerical summarization
print(gdp_data.describe(include= 'all'))
```

```python
#check for missing values and dtypes
print(gdp_data.info())
#there are no missing values 
```

```python

#renaming an individual coulumn 'Life Expectancy at birth (years)'
gdp_data.rename (columns ={'Life expectancy at birth (years)':'life_expectancy'}, inplace=True) 
print(gdp_data.columns)
```

```python
#distinct values for a columns
print(gdp_data.Country.unique())


#there are 6 countries in this data set with info on life expectancy and GDP spanning a 15 year period(2000-2015)
```

```python
# What is the average life expectancy in these nations?

average_life_expectancy = gdp_data.groupby('Country')['life_expectancy'].mean().reset_index()
print(average_life_expectancy)
```

```python
# What is the distribution of that life expectancy?- 0ne quantitative variable (histogram/box plot)
#histogram
# sns.displot(gdp_data.life_expectancy, bins=30, )
# plt.show()
# plt.clf()


#box plot (cummulative)
sns.boxplot(gdp_data.life_expectancy)
plt.title('Distribution of Life Expectancy')
plt.ylabel('Years')
plt.show()
plt.clf()

# boxplot for each country
# for i in range(1,7):
#     country = gdp_data.loc[gdp_data['Country']== i]['life_expectancy']

sns.catplot(x='Country', y='life_expectancy', data=gdp_data, kind='box', height=5, aspect=2)
# Set labels and title
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.title('Distribution of Life Expectancy in Six Countries')

plt.show()
 

#Life_expectancy is between 71-82 years for the majority, with an average of 76 years, however, there are several outliers, especially on the low 
```

```python
# Has life expectancy increased over time in the six nations?
# The facet grid of line graphs mapping Life Expectancy by country

g = sns.FacetGrid(data=gdp_data, col='Country', col_wrap=3, height=4)
g.map(sns.lineplot, 'Year', 'life_expectancy', marker='o')

# Set titles for each subplot
for ax in g.axes:
    ax.set_title(ax.get_title().split('=')[1].strip())

# Adjust spacing between subplots
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Life Expectancy by Country', fontsize=16)
plt.show()
plt.clf()

#yes, but there's a significant increase for zimbabwe from 40 to 65, while the other countries have only about +5yrs
```

```python
# Has GDP increased over time in the six nations?
# The facet grid of line graphs mapping GDP by country

g = sns.FacetGrid(data=gdp_data, col='Country', col_wrap=3, height=4)
g.map(sns.lineplot, 'Year', 'GDP', marker='o')

# Set titles for each subplot
for ax in g.axes:
    ax.set_title(ax.get_title().split('=')[1].strip())

# Adjust spacing between subplots
plt.subplots_adjust(top=0.9)
g.fig.suptitle('GDP by Country', fontsize=16)

# Show the plot
plt.show()
plt.clf()

#China and the USA have significantly increased their GDP, Zimbabwe has the least significant increase in GDP, almost stagnated

```

```python
#Statistical Analysis:

# Conduct statistical tests to quantify the relationship between GDP and life expectancy.

#Is there a correlation between GDP and life expectancy of a country?
#pearson's correlation
from scipy.stats import pearsonr

corr_gdp_Exp, pvalue = pearsonr(gdp_data['GDP'], gdp_data['life_expectancy'])
print(corr_gdp_Exp)

#the cummulative correlation is 0.3432 which indicates a linear relationship

#the correlation between GDP and life expectancy for each country?

# p_values = []
# for i in gdp_data.Country:
#     corr_gdp_Exp, pvalue = pearsonr(gdp_data['GDP'],gdp_data['life_expectancy'])
#     p_values.append(corr_gdp_Exp)
# print(p_values)
```

```python
# Use techniques like regression analysis to model this relationship and assess its significance.
a model to predict life_expectancy based on GDP
import statsmodels.api as sm

model = sm.OLS.from_formula('life_expectancy ~ GDP', data= gdp_data)
results = model.fit()
print(results.params)

# The intercept in the model (7.005478e+01) represents the estimated life expectancy when GDP is Zero. 
# However, this interpretation might not be meaningful in the context of GDP and life expectancy, as GDP cannot be zero 
# Instead, the intercept serves to adjust the baseline life expectancy.

# To determine the significance of the relationship between GDP and life expectancy,
# you would typically look at the p-value associated with the GDP coefficient.
```
